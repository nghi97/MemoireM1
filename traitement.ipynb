```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud,  STOPWORDS
```

```python
import re
text=df['text']
text2=[]
for i in text:
    i=re.sub(r'httpst\S+', '',i)
    text2.append(i)
print(text2)
```

```python

df['text']=text2
df['text']
```

```python
comment_words = ''
stopwords = set(STOPWORDS)
 
# iterate through the csv file
for val in df.text:
     
    # typecaste each val to string
    val = str(val)
 
    # split the value
    tokens = val.split()
     
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
     
    comment_words += " ".join(tokens)+" "
 
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
Stationary=df2.groupby('YMD').agg({'label':'count'})
```

```python
Stationary
```

```python
rolling_mean = Stationary.rolling(7).mean()
rolling_std = Stationary.rolling(7).std()
```

```python
plt.plot(Stationary, color="blue",label="Original Tweets Data")
plt.plot(rolling_std, color="black", label = "Rolling Standard Deviation in Tweets Number")
plt.plot(rolling_mean, color="red", label="Rolling Mean Tweets Number")
plt.title("Tweets Time Series, Rolling Mean, Standard Deviation")
plt.legend(loc="best")
plt.xticks(rotation=45)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adft = adfuller(Stationary,autolag="AIC")
```

```python
output_df = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], adft[4]['1%'], adft[4]['5%'], adft[4]['10%']]  , "Metric":["Test Statistics","p-value","No. of lags used","Number of observations used", 
                                                        "critical value (1%)", "critical value (5%)", "critical value (10%)"]})
print(output_df)
```

```python
autocorrelation_lag1 = Stationary['label'].autocorr(lag=1)
print("One Month Lag: ", autocorrelation_lag1)
```

```python
autocorrelation_lag3 = Stationary['label'].autocorr(lag=3)
print("Three Month Lag: ", autocorrelation_lag3)

autocorrelation_lag6 = Stationary['label'].autocorr(lag=6)
print("Six Month Lag: ", autocorrelation_lag6)

autocorrelation_lag9 = Stationary['label'].autocorr(lag=9)
print("Nine Month Lag: ", autocorrelation_lag9)
```
